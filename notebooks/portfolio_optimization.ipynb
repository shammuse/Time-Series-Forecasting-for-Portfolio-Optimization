## Efficient Frontier in Portfolio Optimization

### Key Concepts

- **Efficient Frontier**: The curve represents the set of portfolios that provide the highest expected return for a given level of risk (volatility).
- **Optimized Portfolio**: The red star marks the portfolio on the efficient frontier that offers the best risk-adjusted return, typically measured by the Sharpe Ratio.

### Interpretation

- **Risk and Return**: Points to the left and below the efficient frontier are suboptimal, meaning they offer either lower returns for the same risk or higher risk for the same return.
- **Optimized Portfolio**: The red star indicates the portfolio that strikes the best balance between risk and return, maximizing the Sharpe Ratio.

### Sharpe Ratio

The color bar on the right represents the **Sharpe Ratio**, a measure of risk-adjusted return. Higher Sharpe Ratios indicate better performance, with the optimized portfolio typically achieving a high Sharpe Ratio.
